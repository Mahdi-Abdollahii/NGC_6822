```python
from AstroDC import AstroDC
```

```python
Z = [0.0001,0.00015,0.0002,0.0003,0.0005,0.0008,
     0.001,0.0015,0.002,0.003,0.006,0.008,
     0.01,0.012]
for i in Z:
  Z_name = ''.join(list(str(i))[2:])
  AstroDC.Stev_Isochrone_CSV('/content/NGC_6822/Data_Isochrones/ISO_'+ Z_name +'_01.dat','ISO_01', Description = False )
  ISO_df_01= pd.read_csv("ISO_01.csv")

  ISO_df_01['logAge'] = round(ISO_df_01['logAge'], 3)
  index  = list(ISO_df_01.groupby(['logAge']).Kmag.idxmin())
  df_n = ISO_df_01.iloc[index]
  plt.plot(df_n['Kmag'],df_n['Mass'])

plt.ylabel('Mass', fontsize = 13, fontname = "serif")
plt.xlabel('K$_s$ [mag]', fontsize = 13, fontname = "serif")
```
